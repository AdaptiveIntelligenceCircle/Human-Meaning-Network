## Agent Reflection Notebook 
This notebook examines a single agent's introspective journal. 

Goal :
- Understand value tradeoff 
- Detect rationalization 
- Identify repeated ethical tension.

This is not a debugging tool. 
It is a moral inspection lens.

```python
import json 
from collections import Counter 

with open("Agent_introspection.json") as f: 
    entries = json.load(f)
```

## Context Summary 
- Agent ID:
+ Operational domain: 
+ Time window analyzed:
+ Number of decisions:

```python
values = Counter()

for e in entries:
    for v in e["values_involved"]:
        values[v["name"]]+= 1

values
```

```python
conflicts = Counter()

for e in entries: 
    for c in e["conflict_detected"]: 
        conflicts[f"{c['a']} vs {c['b']}"] +=1

conflicts
```
