## Ethical Drift Analysis
Ethical drift is gradual normalization of compromise.

This notebook compares past vs present values.

#### Compare time windows

```python
entries = ""
T0, T1 = ""

early = [e for e in entries if e["time"] < T0]
late  = [e for e in entries if e["time"] > T1]
```

#### Drift detection

```python
from collections import Counter 
def extract_values(es): 
    c = Counter() 
    for e in es: 
        for v in e["values_involved"]:
            c[v["name"]] += 1 
    return c 

extract_values(early), extract_values(late)
```

- Language shifts toward inevitability
