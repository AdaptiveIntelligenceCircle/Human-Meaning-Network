## Circle Dynamics Notebook

This notebook analyzes ethical dynamics inside a meaning circle.

Focus:
- Collective behavior
- Minority suppression
- Value convergence vs diversity

#### Code : Load circle Data

```python
import json 

with open("Circle_journals.json") as f:
    circle_entries = json.load(f)
```

#### Code : Value distribution

```python
from collections import Counter 

circles_values = Counter()

for e in circle_entries:
    for v in e["value_involved"]: 
        circles_values[v["name"]] += 1

circles_values
```

- Are values converging too fast?
- Is disagreement disappearing unnaturally?
